```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('analisis_202112161114.csv', sep='\t', dtype={"id_mas":str})
```

```python
df.head()
```

```python
df.tail()
```

```python
table = pd.pivot_table(df, 
                       index = [df['fid'] + '-' + df['id_mas'] + '-' + df['acu'] + '-' + df['tm']  + '-' + df['fecha']],
                       columns=['param'],
                       values=['valor'],
                       aggfunc=np.mean)
```

```python
table
```

```python
table.to_csv('temp.csv',na_rep='NaN')
```
